```python
# Se acostumbra cargar pandas con el alias pd

import pandas as pd
```

```python
felinos = pd.read_csv("/Users/leand/Tarea3/datos/felinos.csv", sep = "\t")
```

```python
felinos
```

```python
# Reducción de columnas
felinos = felinos[["species", "stateProvince", "locality", "eventDate"]]

# Eliminación de fila con valores nulos
felinos = felinos.dropna(how='all')

# Cambio de nombre de las columnas a minúsculas y a nombres más claros
felinos = felinos.rename(columns={"species": "especies",
                           "stateProvince": "provincia",
                           "locality": "localidad",
                           "eventDate": "fecha"})

# Cambio del tipo de datos del campo de fecha
felinos["fecha"] = pd.to_datetime(felinos["fecha"])
```
